```python
# Conteo de Datos Atipicos

# Definir niveles mínimo y máximo para cada columna
nivel_minimo_seats = 1
nivel_maximo_seats = 10

nivel_minimo_km_driven = 0  # Asumiendo que los kilómetros recorridos no pueden ser negativos
nivel_maximo_km_driven = 500000  # establecemos un máximo en 500,000 km

nivel_minimo_year = 2000  # Por ejemplo, si quieres establecer un mínimo en el año 2000
nivel_maximo_year = 2023  # Por ejemplo, si quieres establecer un máximo en el año 2023

# Calcular datos atípicos para cada columna
Total_Atipicos_seats = ((Datos['seats'] < nivel_minimo_seats) | (Datos['seats'] > nivel_maximo_seats)).sum()
Total_Atipicos_km_driven = ((Datos['km_driven'] < nivel_minimo_km_driven) | (Datos['km_driven'] > nivel_maximo_km_driven)).sum()
Total_Atipicos_year = ((Datos['year'] < nivel_minimo_year) | (Datos['year'] > nivel_maximo_year)).sum()

# Imprimir resultados
print("El total de datos atípicos en 'seats' es {}".format(Total_Atipicos_seats))
print("El total de datos atípicos en 'km_driven' es {}".format(Total_Atipicos_km_driven))
print("El total de datos atípicos en 'year' es {}".format(Total_Atipicos_year))

```

```python
# Filtrar Datos Atipicos

# Filtrar 'seats' entre niveles mínimo y máximo
Datos=Datos[(Datos['seats']<nivel_minimo_seats)&(Datos['seats']>nivel_maximo_seats)]
# Filtrar 'km_driven' entre niveles mínimo y máximo
Datos = Datos[(Datos['km_driven'] < nivel_maximo_km_driven) & (Datos['km_driven'] > nivel_minimo_km_driven)]
# Filtrar 'year' entre niveles mínimo y máximo
Datos = Datos[(Datos['year'] < nivel_maximo_year) & (Datos['year'] > nivel_minimo_year)]
```

```python
#Validar si se removieron los Datos atipicos

Total_Atipicos=((Datos['seats']<nivel_minimo_seats)|(Datos['seats']>nivel_maximo_seats)).sum()
print("el total de datos atipicos es {}".format(Total_Atipicos))

Total_Atipicos=((Datos['km_driven']<nivel_minimo_km_driven)|(Datos['km_driven']>nivel_maximo_km_driven)).sum()
print("el total de datos atipicos es {}".format(Total_Atipicos))

Total_Atipicos=((Datos['year']<nivel_minimo_year)|(Datos['year']>nivel_maximo_year)).sum()
print("el total de datos atipicos es {}".format(Total_Atipicos))
```

```python
#Porcentaje de datos para procesamiento (train)= 75%, y para prueba (test) = 25%

from sklearn.model_selection import train_test_split

# Se crean variables X y Y para almacenar la información del entrenamiento
X = Datos.drop('selling_price', axis=1)  # Se cargan los datos de entrenamiento sin la columna 'selling_price'
Y = Datos['selling_price']  # Se cargan los datos de entrenamiento de la columna 'selling_price'

# La función train_test_split crea una división de un conjunto de datos en dos bloques: uno de entrenamiento y otro de prueba (train and test)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=0)
```

```python
print (X_train)
```

```python
print (Y_train)
```

```python


#Punto 5 - Entrenar el modelo configurando los diferentes hiperparámetros.

```

```python
# Construir modelo de regresion lineal 

# Construir modelo de regresión lineal
modelo_regresion = LinearRegression()

# Entrenar el modelo
modelo_regresion.fit (X_train, Y_train)

# Realizar predicciones
Prediccion = modelo_regresion.predict(X_test)
```

```python
print (Y_test)
```

```python
print (Prediccion)
```

```python


#Punto 6 - Evaluar el desempeño del modelo en el conjunto de Test con métricas como precisión, recall, F1-score, etc

```

```python
# Desempeño del modelo de regresion

print(classification_report(Y_test,Prediccion))
```

```python


#Punto 7 - Realizar las diferentes gráficas que permitan visualizar los resultados del modelo.

```

```python
# Matrix de confusion de regresion

cm=confusion_matrix(Y_test,Prediccion)
disp=metrics.ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
```
